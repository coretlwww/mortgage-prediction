```python
def create_chart(cols, chart_title):
    fig, axes = plt.subplots(3, 3, figsize=(25, 12))  # Увеличили высоту до 15, чтобы графики не слипались
    axes = axes.flatten()  # Делаем массив осей плоским: от 0 до 8

    # Итерируемся по колонкам (без -1, чтобы не терять последнюю)
    for i, col in enumerate(cols):
        if i < len(axes):  # Защита на случай, если колонок больше, чем ячеек (9)
            chart_title(data=mortgage_data, x=col, ax=axes[i], color='lightgreen')
            if col == "Job":
                axes[i].tick_params(axis="x", labelrotation=45)
            if cols == categorical_cols:
                for container in axes[i].containers:
                    axes[i].bar_label(
                        container,
                        label_type="center",  # Располагает текст ровно по центру столбика
                        color="black",  
                        fontweight="bold",  # Жирный шрифт, чтобы цифры выделялись
                        fontsize=10,  # Размер шрифта
                    )

    # Удаляем пустые (лишние) графики, если колонок оказалось меньше 9
    for j in range(len(cols), len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()  # Автоматически настраивает отступы между графиками
    plt.show()
```

# Making predictions

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor
from numpy import random
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
```

```python
X = mortgage_data.drop('Max Loan Amount (USD)', axis=1)
X.head()
```

```python
X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.25, random_state=1)
```

```python
object_cols = [col for col in X_train.columns if X_train[col].dtype == 'str']
low_cardinality_cols = [col for col in object_cols if X_train[col].nunique() <= 15]
high_cardinality_cols = list(set(object_cols) - set(low_cardinality_cols))
X_train.drop(high_cardinality_cols, axis=1)
X_val.drop(high_cardinality_cols, axis=1)
X_test.drop(high_cardinality_cols, axis=1)
```

```python
categorical_transformer = Pipeline(
    steps = [('onehot', OneHotEncoder(handle_unknown='ignore'))]
)

preprocessor = ColumnTransformer(
    transformers=[('cat', categorical_transformer, low_cardinality_cols)]
)
```

```python
def get_mae(model, train_X, train_y, val_X, val_y):
    new_pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('model', model)
    ])
    new_pipeline.fit(train_X, train_y)
    preds_val = new_pipeline.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return mae
```

```python
random.seed(42)

leaf_size = random.randint(1, 500, 10).tolist()
leaf_size
```

```python
def find_best_param(model_class, param_name, params_list, X_tr, y_tr, X_v, y_v, **extra_kwargs):
    """
    params_list — список значений для перебора (например, n_estimators).
    **extra_kwargs — любые дополнительные фиксированные параметры модели.
    """
    best_mae = float("inf")
    best_param = None

    for value in params_list:
        # 1. Берем дополнительные параметры (например, learning_rate, max_depth)
        model_kwargs = extra_kwargs.copy()
        
        # 2. Добавляем/обновляем перебираемый параметр и random_state
        model_kwargs[param_name] = value
        if "random_state" not in model_kwargs:
            model_kwargs["random_state"] = 0
            
        # 3. Создаем модель
        temp_model = model_class(**model_kwargs)
        
        # 4. Считаем ошибку
        mae = get_mae(temp_model, X_tr, y_tr, X_v, y_v)
        print(f"Parameter {param_name} = {value:<4} | MAE: {mae:.4f}")

        if mae < best_mae:
            best_mae = mae
            best_param = value

    print(f"\nBest {param_name}: {best_param} (Best MAE: {best_mae:.4f})")
    return best_param
```

```python
find_best_param(DecisionTreeRegressor, 'max_leaf_nodes', leaf_size, X_train, y_train, X_val, y_val)
```

```python
random.seed(42)

trees_number = random.randint(15, 400, 10).tolist()
trees_number
```

```python
find_best_param(RandomForestRegressor, 'n_estimators', trees_number, X_train, y_train, X_val, y_val)
```

```python
random.seed(42)

estimators_list = random.randint(100, 1000, 10).tolist()
estimators_list
```

```python
find_best_param(XGBRegressor, 'n_estimators', estimators_list, X_train, y_train, X_val, y_val, learning_rate=0.05, n_jobs=4)
```

```python
model_result = {}
model_result['Decision Tree'] = get_mae(DecisionTreeRegressor(max_leaf_nodes=72, random_state=0), X_train, y_train, X_val, y_val)
model_result['Random Forest'] = get_mae(RandomForestRegressor(n_estimators=285, random_state=0), X_train, y_train, X_val, y_val)
model_result['XGBRegressor'] = get_mae(XGBRegressor(n_estimators=120, learning_rate=0.05, n_jobs=4, random_state=0), X_train, y_train, X_val, y_val)
best_model = min(model_result, key=model_result.get)
best_model
```
