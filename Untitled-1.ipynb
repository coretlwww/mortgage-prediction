```python
# Define a reusable function to plot multiple feature distributions in a 3x3 grid
def create_chart(cols, chart_title):
    fig, axes = plt.subplots(3, 3, figsize=(25, 12))  
    axes = axes.flatten()  # Flatten the 2D grid array into 1D for easy iteration

    for i, col in enumerate(cols):
        if i < len(axes): 
            # Call the plotting function (passed via chart_title argument) for the current column
            chart_title(data=mortgage_data, x=col, ax=axes[i], color='lightgreen')
            
            # Rotate long category labels on the Job axis to prevent overlapping
            if col == "Job":
                axes[i].tick_params(axis="x", labelrotation=45)
                
            # Add text labels inside bars specifically for categorical columns
            if cols == categorical_cols:
                for container in axes[i].containers:
                    axes[i].bar_label(
                        container,
                        label_type="center",  
                        color="black",  
                        fontweight="bold",  
                        fontsize=10, 
                    )

    # Remove empty, unused subplots from the grid if features are fewer than 9
    for j in range(len(cols), len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()  # Adjust spacing to avoid overlapping text and axes labels
    plt.show()
```

# Model Training & Final Predictions

```python
# Import all the necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor
from numpy import random
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
```

```python
# Create the feature matrix X by dropping the target column
X = mortgage_data.drop('Max Loan Amount (USD)', axis=1)
# Display the first 5 rows
X.head()
```

```python
# Perform a two-stage split to create training (60%), validation (20%), and testing (20%) datasets
X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.25, random_state=1)
```

```python
# Identify string columns and separate them by cardinality to manage high-dimensional categorical features
object_cols = [col for col in X_train.columns if X_train[col].dtype == 'str']
low_cardinality_cols = [col for col in object_cols if X_train[col].nunique() <= 15]
high_cardinality_cols = list(set(object_cols) - set(low_cardinality_cols))

# Drop high-cardinality features across all splits to prevent overfitting and excessive feature expansion
X_train.drop(high_cardinality_cols, axis=1)
X_val.drop(high_cardinality_cols, axis=1)
X_test.drop(high_cardinality_cols, axis=1)
```

```python
# Create a preprocessing pipeline to handle categorical features and encode them cleanly
categorical_transformer = Pipeline(
    steps = [('onehot', OneHotEncoder(handle_unknown='ignore'))]
)

# Bundle the transformation step to apply it exclusively to low-cardinality features
preprocessor = ColumnTransformer(
    transformers=[('cat', categorical_transformer, low_cardinality_cols)]
)
```

```python
# Define a helper function to fit a pipeline and evaluate a model's Mean Absolute Error (MAE)
def get_mae(model, train_X, train_y, val_X, val_y):
    # Combine the preprocessor and the given model into a single executable pipeline
    new_pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('model', model)
    ])
    
    # Train the complete pipeline on the training dataset
    new_pipeline.fit(train_X, train_y)
    
    # Generate predictions on the validation set and compute the absolute error metric
    preds_val = new_pipeline.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    
    return mae
```

```python
# Generate a reproducible random list of max_leaf_nodes to hyperparameter tune the model
random.seed(42)

leaf_size = random.randint(1, 500, 10).tolist()
leaf_size
```

```python
# Define a reusable function to tune any model parameter and plot the resulting MAE curve
def plot_param_vs_mae(model_class, param_name, params_list, X_tr, y_tr, X_v, y_v, **extra_kwargs):
    # Enure the parameter values are unique and sorted from smallest to largest for a clean line plot
    sorted_params = sorted(list(set(params_list)))
    mae_scores = []
    
    # Loop through each parameter value to collect validation errors
    for value in sorted_params:
        model_kwargs = extra_kwargs.copy()
        model_kwargs[param_name] = value
        if "random_state" not in model_kwargs:
            model_kwargs["random_state"] = 0
            
        # Instantiate model, run your existing pipeline function, and save the score
        temp_model = model_class(**model_kwargs)
        mae = get_mae(temp_model, X_tr, y_tr, X_v, y_v)
        mae_scores.append(mae)
        print(f"Parameter {param_name} = {value:<4} | MAE: {mae:.4f}")

    # Set up and style the performance visualization chart
    plt.figure(figsize=(6, 5))
    sns.lineplot(x=sorted_params, y=mae_scores, marker='o', color='lightgreen', linewidth=2.5)
    
    # Dynamically label the chart based on the passed argument names
    model_name = model_class.__name__
    plt.title(f'{model_name}: Validation MAE vs. {param_name}', fontsize=14, fontweight='bold')
    plt.xlabel(f'{param_name}', fontsize=12)
    plt.ylabel('Mean Absolute Error (MAE)', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.6)
    
    plt.show()
    
    # Return the best value programmatically just like your previous function did
    best_mae_index = mae_scores.index(min(mae_scores))
    return sorted_params[best_mae_index]
```

```python
# Tune and plot the MAE curve for DecisionTreeRegressor to find the optimal max_leaf_nodes
plot_param_vs_mae(DecisionTreeRegressor, 'max_leaf_nodes', leaf_size, X_train, y_train, X_val, y_val)
```

The hyperparameter tuning results show that setting max_leaf_nodes to 72 minimizes the validation Mean Absolute Error (MAE) at $234,277.65. Larger node sizes (like 436) cause the error to increase due to overfitting, while smaller settings (like 21) restrict the tree too much, making 72 the sweet spot for balancing model complexity and generalization.

```python
# Generate a reproducible random list of n_estimators (tree counts) to tune Random Forest model
random.seed(42)

trees_number = random.randint(15, 400, 10).tolist()
trees_number
```

```python
# Execute the hyperparameter tuning function to find the optimal number of trees for a Random Forest
plot_param_vs_mae(RandomForestRegressor, 'n_estimators', trees_number, X_train, y_train, X_val, y_val)
```

The hyperparameter tuning results identify 285 as the optimal number of trees (n_estimators), minimizing the Mean Absolute Error (MAE) at $235,565.97. Increasing the forest size beyond this point provides diminishing returns and slight overfitting, while fewer trees lead to higher variance and less stable predictions.

```python
# Generate a reproducible random list of estimator counts to hyperparameter tune an ensemble model
random.seed(42)

estimators_list = random.randint(100, 1000, 10).tolist()
estimators_list
```

```python
# Execute the hyperparameter tuning function to find the optimal n_estimators for an XGBoost Regressor
plot_param_vs_mae(XGBRegressor, 'n_estimators', estimators_list, X_train, y_train, X_val, y_val, learning_rate=0.05, n_jobs=4)
```

The hyperparameter tuning results show that a smaller forest size of 120 trees provides the best generalization, minimizing the Mean Absolute Error (MAE) at $234,416.91. Adding more estimators steadily degrades performance—with error peaking near 960 trees—confirming that larger ensembles are overfitting the training noise without adding predictive value.

```python
# Collect validation MAE scores for all optimized models and identify the best-performing algorithm
model_result = {}
model_result['Decision Tree'] = get_mae(DecisionTreeRegressor(max_leaf_nodes=72, random_state=0), X_train, y_train, X_val, y_val)
model_result['Random Forest'] = get_mae(RandomForestRegressor(n_estimators=285, random_state=0), X_train, y_train, X_val, y_val)
model_result['XGBRegressor'] = get_mae(XGBRegressor(n_estimators=120, learning_rate=0.05, n_jobs=4, random_state=0), X_train, y_train, X_val, y_val)

# Determine the model with the minimum Mean Absolute Error
best_model = min(model_result, key=model_result.get)
best_model
```

The model comparison reveals that the tuned Decision Tree out-performs the ensemble methods, achieving the lowest validation Mean Absolute Error. This outcome strongly implies that the dataset contains distinct, rigid decision boundaries—likely due to the fixed bank policies and data capping identified during data exploration—which a single, optimized tree can pinpoint perfectly without the excessive smoothing or overfitting introduced by the Random Forest and XGBoost models.

```python
# Construct the final pipeline with the chosen model and evaluate it on the held-out test data
new_pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('model', DecisionTreeRegressor(max_leaf_nodes=72, random_state=0))
    ])

# Fit the model on the full training partition and generate test predictions
new_pipeline.fit(X_train, y_train)
predictions = new_pipeline.predict(X_test)
predictions
```

```python
plt.figure(figsize=(6, 5))

# Plot training labels distribution vs the test predictions distribution
sns.kdeplot(y_train, label='Training Data (Actual)', color='blue', fill=True, alpha=0.3)
sns.kdeplot(predictions, label='Test Data (Predicted)', color='lightgreen', fill=True, alpha=0.3)

plt.title('Sanity Check: Training Actuals vs. Test Predictions', fontsize=14, fontweight='bold')
plt.xlabel('Max Loan Amount (USD)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend()
plt.show()
```

The sanity check plot reveals severe underfitting and variance compression in the final Decision Tree model. While a maximum of 72 leaf nodes minimized the Mean Absolute Error (MAE), it forced the predictions (green curve) into a narrow band between $400,000 and $1,000,000, completely failing to capture the true, wide distribution of the actual loan amounts (blue curve).

Limitations of the Study
- Artificial Data Truncation (Right Censoring): Severe capping spikes at 50 Employment Years and an 850 Credit Score mask the true variance of top-tier applicants, restricting the model's ability to learn from elite borrower profiles.
- Severe Target Skewness: The heavy right-hand tail and extreme upper-bound outliers in Max Loan Amount (USD) introduce mathematical bias, distorting standard error metrics and diminishing predictive accuracy for high-value loans.
- Model Underfitting and Variance Compression: Restricting the winning Decision Tree to 72 leaf nodes compresses predictions into a narrow, centralized band ($400,000 to $1,000,000), completely failing to capture the actual lower-bound and upper-bound market extremes.
- Institutional Policy Rigidities: Features like Interest Rate operate on strict, discrete corporate pricing tiers rather than fluid market dynamics, limiting the model's generalizability to other financial institutions. 
- Potential Synthetic Sampling Bias: Perfectly balanced features like Gender and uniform distributions like Age suggest an engineered or heavily curated dataset, which may not accurately reflect real-world, unstratified credit applicant populations.
