```python
mortgage_data['Max Loan Amount (USD)'] = mortgage_data['Max Loan Amount (USD)'].astype('int64')
y = mortgage_data['Max Loan Amount (USD)']
y
```

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 5))

sns.boxplot(data=mortgage_data, x=y, ax=axes[0], color='lightgreen')
axes[0].ticklabel_format(style='plain', axis='x')
axes[0].set_title('Диаграмма размаха (Boxplot)')

sns.histplot(data=mortgage_data, x=y, ax=axes[1], color='lightgreen')
axes[1].ticklabel_format(style='plain', axis='x')
axes[1].set_title('Гисторамма (Histogram)')

plt.show()
```

```python
y.describe().round(0)
```

```python
# 1. Находим 25-й (Q1) и 75-й (Q3) перцентили
q1 = y.quantile(0.25)
q3 = y.quantile(0.75)

# 2. Считаем межквартильный размах (IQR)
iqr = q3 - q1

# 3. Вычисляем теоретические границы усов
upper_whisker = q3 + 1.5 * iqr
```

```python
y.where(y > upper_whisker).count().item()
```

```python
numeric_cols = mortgage_data.select_dtypes(include=['int64', 'float64']).columns.tolist()
numeric_cols
```

```python
def create_chart(chart_title):
    fig, axes = plt.subplots(3, 3, figsize=(20, 10))  # Увеличили высоту до 15, чтобы графики не слипались
    axes = axes.flatten()  # Делаем массив осей плоским: от 0 до 8

    # Итерируемся по колонкам (без -1, чтобы не терять последнюю)
    for i, col in enumerate(numeric_cols):
        if i < len(axes):  # Защита на случай, если колонок больше, чем ячеек (9)
            chart_title(data=mortgage_data, x=col, ax=axes[i], color='lightgreen')

    # Удаляем пустые (лишние) графики, если колонок оказалось меньше 9
    for j in range(len(numeric_cols), len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()  # Автоматически настраивает отступы между графиками
    plt.show()
```

```python
create_chart(sns.histplot)
```

```python
create_chart(sns.boxplot)
```
